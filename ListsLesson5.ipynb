```python
#this allows me to see all the columns, without it I'd only see some of them because this is a big data set
pd.set_option('display.max_columns', None)
#This data was seperated by |, the encoding uses the ISO and the data type is object
df = pd.read_csv(r'C:\Users\efann\PROPERTY TAX\TarrantPropData2020.txt', sep ="|", encoding='ISO-8859-1', dtype='object')
#View the first 5 rows of data, I could put a number in the parentheses and see that many rows otherwise automatic blank is 5 rows
df.head()
```

```python
#This is a blank list called propertytax
propertytax_list = []

#Looping within this Total_Value column, the : without a number means look at all rows and the : at the end tells python to enter into that function
for index, row in df.loc[:, 'Total_Value'].iteritems():
    propertytax = (float(row) * 2.10)/100
    propertytax_list.append(propertytax)
#the propertytax columm this creates will be a float row ie has decimal values and is the Total_value times the tax rate of 2.10 then divide by 100 bc it's a percentage
#the information is then placed in the blank propertytax list from the start of the cell which takes the data outside of the loop, making it easier to use elsewhere

#This adds the list above as a new column in the original dataframe
df['Property_Tax'] = pd.Series(propertytax_list)

#lets us look at the first five entries to make sure it worked
df.head()
```
