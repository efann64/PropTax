```python
situs_address = df['Situs_Address'].str.split(expand = True)

Directions = ['N', 'E', 'W', 'S']

#these are empty lists
Street_Type=[]
Street_Name = []
Street_No = []
Street_Address = []
Prefix = []
Suffix = []


for row in situs_address.loc[:].values.tolist():
#    print(row)
    
    #4.a.a this will pull addressess that are FOUR words long, have a direction in the second column and have ST as the last word    
    if row[0] is not None and row[1] is not None and row[2] is not None and row[3] is not None and str(row[0]).isnumeric() == True and row[1] in Directions and row[3] == 'ST':      
        Street_Type.append(row[3])
        Street_Name.append(row[1] + ' ' + row[2]) 
        Street_No.append(row[0])
        Street_Address.append(row[0] + ' ' + row[1] + ' ' + row[2] + ' ' + row[3])
        Prefix.append(row[1])
        Suffix.append(None)

    #There are no addressess that are four words long and do NOT have a direction in the second column
    #if row[0] is not None and row[1] is not None and row[2] is not None and row[3] is not None and str(row[0]).isnumeric() == True and row[1] not in Directions and row[3] == 'ST': 


    #4.a.a this will pull addressess that are THREE words long, and have ST as the last word
    elif row[0] is not None and row[1] is not None and row[2] is not None and row[3] is None and str(row[0]).isnumeric() == True and row[2] == 'ST':      

        Street_Type.append(row[2])
        Street_Name.append(row[1] + ' ' + row[2]) 
        Street_No.append(row[0])
        Street_Address.append(row[0] + ' ' + row[1] + ' ' + row[2])
        Prefix.append(None)
        Suffix.append(None)

#originally I had print(row) above and it gave me all of the items I specified I was looking for- three worded addresses with ST as the last word.  When I do not have row in the parenthasis, I get no data, which may mean there are no missing rows of data that are not captures in the code above, need to confirm

#adding new columns to our dataframe to accommodate the new data we just split and cleaned
df['Street_Type'] = pd.Series(Street_Type)
df['Street_Name'] = pd.Series(Street_Name)
df['Street_No'] = pd.Series(Street_No)
df['Street_Address'] = pd.Series(Street_Address)
df['Prefix'] = pd.Series(Prefix)
df['Suffix'] = pd.Series(Suffix)

#checking to make sure our new columns that contain our situs_address transformation data shows up; unsure if I should have the prefix in the street address if not I need to remove that row addition from Street_Name
df.head()
```
